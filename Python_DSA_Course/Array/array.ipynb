```python
# array/list : the values are stored on addresses(hexadecimal numbers) on RAM in continuous memory locations
# analogous to ArrayList in Java
stock_prices=[345,456,421,235,690]
day4_price= stock_prices[3] # Accessing element: big O = O[1]
print(day4_price)
```

```python
# check on which day the price falls
for i in range (len(stock_prices)):
    if stock_prices[i+1]<stock_prices[i]:
        print("On the day {} the price is falling, time to sell!".format(i+2)) # big O = O[n]
        break
```

```python
stock_prices.append(989) # To add an item to the end of the list, use the append() method
stock_prices.insert(2,789) # requires position and value
stock_prices.remove(789) # The remove() method removes the specified item.
stock_prices.pop(3) # The pop() method removes the specified index.
del stock_prices[0] # The del keyword also removes the specified index. All of the above have big O= O[n]
print(stock_prices)
```

```python
# List in python are dynamic arrays. Java and C++ have both dynamic and static arrays
# Let us say your expense for every month are listed below,
# January - 2200
# February - 2350
# March - 2600
# April - 2130
# May - 2190

# 1. In Feb, how many dollars you spent extra compare to January?
# 2. Find out your total expense in first quarter (first three months) of the year.
# 3. Find out if you spent exactly 2000 dollars in any month
# 4. June month just finished and your expense is 1980 dollar. Add this item to our monthly expense list
# 5. You returned an item that you bought in a month of April and got a refund of 200$. Make a correction to your monthly expense list based on this
months = ["January","February", "March", "April", "May"]
expenses= [2200, 2350, 2600, 2130, 2190]
extra = expenses[1]-expenses[0]
print(f"extra: {extra}")
tot_3mon = expenses[0]+expenses[1]+expenses[2]
print(f"total for 3 months: {tot_3mon}")
for i in range(len(expenses)):
    if expenses[i]==2000:
        print("2000$ in the month of {}".format(months[i]))
        break
else:
    print("No such month")

# Or we could simply use the following code for q3, just for checking and not for printing the month:
if 2000 in expenses:
    print ("Yes")
else:
    print("No")

months.append("June")
print(months)
expenses.append(1980)
print(expenses)
expenses[3]=expenses[3]-200
print(expenses)
```

```python
# 1. Length of the list
# 2. Add 'black panther' at the end of this list
# 3. You realize that you need to add 'black panther' after 'hulk',
#    so remove it from the list first and then add it after 'hulk'
# 4. Now you don't like thor and hulk because they get angry easily :)
#    So you want to remove thor and hulk from list and replace them with doctor strange (because he is cool).
#    Do that with one line of code.
# 5. Sort the heros list in alphabetical order (Hint. Use dir() functions to list down all functions available in list)

heroes=['spider man','thor','hulk','iron man','captain america']
print(len(heroes))
heroes.append("black panther")
print(heroes)
heroes.remove("black panther")
heroes.insert(3,"black panther")
print(heroes)
heroes=heroes[:1]+heroes[3:5]+["doctor strange"] # list slicing: heroes[a:b] means a to b where a is included and b is not
print(heroes)

# The sort() method sorts the list ascending by default; it sorts internally, no need of new list. list.sort(reverse=True|False, key=myFunc)
heroes.sort()
print(heroes)
```
